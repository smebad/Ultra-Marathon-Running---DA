## Importing Libraries

```python
import pandas as pd
import seaborn as sns
```

## Loading the Data

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
# Display the first 10 rows of the dataframe
df.head(10)
```

```python
# Checking the shape of the dataframe
df.shape
```

```python
# Checking the data types of each column
df.dtypes
```

## Data Cleaning

```python
# Filtering the dataframe for events with distance of '50km'
df[df['Event distance/length'] == '50km']
```

```python
# Combining filters for 50km and 50mi events
df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
# Further filtering for events that occurred in the year 2020
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

```python
# Extracting the country information from the event name
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
# Filtering the dataframe for events in the USA
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
# Combining all the filters together: events with distance '50km' or '50mi', occurred in 2020, and in the USA
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
# Display the first 10 rows of the filtered dataframe
df2.head(10)
```

```python
# Checking the shape of the filtered dataframe
df2.shape
```

```python
# Removing the country information from the event name
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
df2.head()
```

```python
# Cleaning up the age column by calculating the age of athletes
df2['athelete age'] = 2020 - df2['Athlete year of birth']
```

```python
# Removing the 'h' from Athlete performance column
df2['Athlete performance'] = df2['Athlete performance'].str.replace('h', '')
df2.head()
```

```python
# Dropping unnecessary columns
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
df2.head()
```

```python
# Checking for null values
df2.isna().sum()
```

```python
# Dropping rows with any null values
df2 = df2.dropna()
df2.shape
```

```python
# Checking for duplicates
df2.duplicated().sum()
```

```python
# Resetting the index of the dataframe
df2.reset_index(drop=True, inplace=True)
```

```python
# Checking data types of the dataframe
df2.dtypes
```

```python
# Converting 'athelete age' to integer type
df2['athelete age'] = df2['athelete age'].astype(int)
```

```python
# Converting 'Athlete average speed' to float type
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
df2.dtypes
df2.head()
```

## Renaming Columns

```python
# Renaming columns for better readability
df2 = df2.rename(columns={'Year of event': 'year',
                          'Event dates': 'marathon day',
                          'Event name': 'marathon name',
                          'Event distance/length': 'distance',
                          'Event number of finishers': 'finishers',
                          'Athlete performance': 'performance',
                          'Athlete gender': 'gender',
                          'Athlete average speed': 'avg_speed',
                          'Athlete ID': 'id'})

df2.head()
```

```python
# Reordering columns for better organization
df3 = df2[['marathon day', 'marathon name', 'distance', 'finishers', 'performance', 'gender', 'avg_speed', 'id', 'athelete age']]
df3.head()
```

## Data Visualization

```python
# Distribution of event distances
sns.histplot(df3['distance'])
```

```python
# Distribution of event distances by gender
sns.histplot(df3, x='distance', hue='gender')
```

```python
# Distribution of average speed for 50km events
sns.displot(df3[df3['distance'] == '50km']['avg_speed'])
```

```python
# Violin plot of average speed by distance and gender
sns.violinplot(data=df3, x='distance', y='avg_speed', hue='gender', split=True, inner='quartile')
```

```python
# Linear model plot of athlete age and average speed by gender
sns.lmplot(data=df3, x='athelete age', y='avg_speed', hue='gender')
```

## Analysis

```python
# Calculating the difference between average speed of male and female athletes for the 50km and 50mi race
df3.groupby(['distance', 'gender'])['avg_speed'].mean()
```

```python
# Age group analysis for the 50mi race
df3.query('distance == "50mi"').groupby('athelete age')['avg_speed'].agg(['mean', 'count']).sort_values(by='mean', ascending=False).query('count > 10')
```
